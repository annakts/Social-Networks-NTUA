```python
%matplotlib inline
import matplotlib
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
```

### Α) Κατασκευή και οπτικοποίηση σύνθετων τύπων δικτύου

### Τοπολογία REG για βαθμό πλέγματος $d = 4$

```python
random.seed(seedno)
#n=4x μέγεθος δικτύου: αντικαταστήστε το x με το τελευταίο ψηφίο του αριθμού μητρώου σας.
n=49
d=4
# για την κατασκευή του REG χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
# Επιλέξτε προσεκτικά την τιμή της παραμέτρου p. 
REG=nx.watts_strogatz_graph(n, d, 0.0) 
REG.graph['name'] = 'REG'

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(REG, node_size=300, node_color="orange", with_labels=True)
```

### Τοπολογία RG (ER) για αριθμό συνδέσεων $M = 150$

```python
random.seed(seedno)
#n=4x 
n=49
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER=nx.gnm_random_graph(n, M)
RGER.graph['name'] = 'RGER' 

# οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(RGER, node_size=250, node_color="orange", with_labels=True)
```

### Τοπολογία RGG  με ακτίνα σύνδεσης $R=0.35$

```python
random.seed(seedno)
#n=4x
n=49
R=0.35

# για την κατασκευή του RGG χρησιμοποιήστε την random_geometric_graph από το networkx
RGG=nx.random_geometric_graph(n, R, 2)
RGG.graph['name'] = 'RGG'

#οπτικοποίηση: https://networkx.org/documentation/stable/auto_examples/drawing/plot_random_geometric_graph.html
pos = nx.get_node_attributes(RGG, "pos")
dmin = 1
ncenter = 0

for n in pos:
    x, y = pos[n]
    d = (x - 0.5) ** 2 + (y - 0.5) ** 2
    if d < dmin:
        ncenter = n
        dmin = d
# color by path length from node near center
p = dict(nx.single_source_shortest_path_length(RGG, ncenter))

labels={}
for n in RGG.nodes():
  labels[n] = n

plt.figure(figsize=(8, 8))
nx.draw_networkx_edges(RGG, pos, alpha=0.4)
nx.draw_networkx_nodes(
    RGG,
    pos,
    nodelist=list(p.keys()),
    node_size=250,
    node_color=list(p.values()),
    cmap=plt.cm.Reds_r,
)
nx.draw_networkx_labels(RGG, pos, labels, font_size=10, font_color="white")

plt.xlim(-0.05, 1.05)
plt.ylim(-0.05, 1.05)
plt.axis("off")
plt.show()

```

### Τοπολογία SF με παράμετρο επισύναψης $d=4$

```python
random.seed(seedno)
#n=4x 
n=49
d=4

# για την κατασκευή του SF χρησιμοποιήστε την barabasi_albert_graph από το networkx
SF= nx.barabasi_albert_graph(n, d)
SF.graph['name'] = 'SF'

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SF, node_size=250, node_color="orange", with_labels=True)
```

### Τοπολογία SW με βαθμό αρχικού πλέγματος $d=4$ και πιθανότητα ανασύνδεσης $p=0.25$

```python
random.seed(seedno)
#n=4x
n=49
d=4
p=0.25

# για την κατασκευή του SW χρησιμοποιήστε την watts_strogatz_graph από το networkx. 
SW= nx.watts_strogatz_graph(n, d, p) 
SW.graph['name'] = 'SW' 

#οπτικοποίηση: χρησιμοποιήστε την draw_circular από το networkx
nx.draw_circular(SW, node_size=250, node_color="orange", with_labels=True)
```

### B) Μετρικές δικτύου

### B1. Συντελεστής ομαδοποίησης (Clustering coefficient)

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β1.1 Να υπολογιστεί ο συντελεστής ομαδοποίησης κάθε κόμβου και να αναπαρασταθεί η κατανομή του.
#### Β1.2  Να υπολογιστεί ο μέσος συντελεστής ομαδοποίησης και η διασπορά του.
#### Β.1.3 Να βρεθούν οι κόμβοι με το μικρότερο και το μεγαλύτερο συντελεστή ομαδοποίησης. 
#### Να σχολιαστούν τα αποτελέσματα.

```python
Graphs = [REG, RGER, RGG, SF, SW]
```

```python
#B1.1
def cc(G, name):
    for n in G:
        G.nodes[n]['cc'] = nx.clustering(G, n)
    
    plt.figure()
    plt.hist([x[1] for x in G.nodes(data='cc')])
    plt.title(f'Clustering Coefficient Distribution for {name}')
    plt.xlabel("Clustering Coefficient")
    plt.ylabel("Frequency")
    plt.show()
    
for g in Graphs :
    cc(g, g.graph['name'])
```

```python
#B1.2
print('Average Clustering Coefficient')
for g in Graphs :
  print(g.graph['name'] + ' : Mean : ' + str(round(nx.average_clustering(g),4)) + ', Variance : ' + str(round(np.var([x[1] for x in g.nodes(data='cc')]),4)))
```

```python
#B1.3
def minV(G):
    # min_cc = min(list(G.nodes(data='cc')), key=lambda x : x[1])
    min_cc = min(x[1] for x in G.nodes(data='cc'))
    return (min_cc, [x[0] for x in list(G.nodes(data='cc')) if x[1]==min_cc])

def maxV(G):
    # max_cc = max(list(G.nodes(data='cc')), key=lambda x : x[1])
    max_cc = max(x[1] for x in G.nodes(data='cc'))
    return (max_cc, [x[0] for x in list(G.nodes(data='cc')) if x[1]==max_cc])

print('Vertices with MIN & MAX Clustering Coefficient')
for g in Graphs :
  print(g.graph['name'] + ' -> MIN = ' + str(round(minV(g)[0], 2)) + ' : ' + str(minV(g)[1]) + ', ΜΑΧ = ' + str(round(maxV(g)[0], 2)) + ': ' + str(maxV(g)[1]))
```

Σχολιασμός Αποτελεσμάτων:

REG: Παρατηρούμε ότι όλοι οι κόμβοι έχουν τον ίδιο συντελεστή ομαδοποίησης (=0.5) , διότι υπάρχει ομοιομορφία στα χαρακτηριστικά τους.

RGER: Παρατηρούμε ότι οι περισσότεροι συντελεστές ομαδοποίησης είναι αρκετά χαμηλή και ομοίως κατά συνέπεια και ο μέσος συντελεστής ομαδοποίησης. Συνεπώς συμπεραίνουμε ότι είναι δεν υπάρχουν πολλά clusters στην τοπολογία, κάτι που οφείλεται κατά πάσα πιθανότητα στην τυχαιότητα των ακμών.

RGG: Εδώ παρατηρούνται οι μεγαλύτεροι συντελεστές ομαδοποίησης και ο υψηλότερος μέσος, γεγονός που σημαίνει ότι είναι η τοπολογία με τα περισσότερα clusters.

SF: Σε αυτή την τοπολογία οι περισσότεροι κόμβοι έχουν σχετικά μικρούς συντελεστές ομαδοποίησης, ενώ λίγοι έχουν αρκετά υψηλούς με αποτέλεσμα να ανεβάζουν τη μέση τιμή. Η δομή αυτή παρομοιάζει πιο ρεαλιστικά ένα πραγματικό κοινωνικό δίκτυο. 

SW: Όπως και στην Scale-free η πλειοψηφία των κόμβων έχει μικρούς συντελεστές, ενώ λίγοι κόμβοι έχουν συντελεστές ομαδοποιήσης με αρκετά μεγάλες τιμές. Σε σύγκριση με την SF τοπολογία εδώ η το εύρος της διαφοράς στους χαμηλούς και υψηλούς συντελεστές είναι μεγαλύτερο, όπως και στο πλήθος των κόμβων με υψηλές και χαμηλές τιμές συντελεστή. Για το λόγο αυτό η Small-World τοπολογία αποτελεί ακόμα καλύτερη αναπαράσταση ενός πραγματικού δικτύου.

### Β2. Μέσο μήκος μονοπατιού - Διάμετρος - Ακτίνα γράφου

#### Για κάθε μια από τις τοπολογίες REG, RG(ER), RGG, SF(BA) και SW(WS):
#### Β2.1 Nα υπολογιστεί το μέσο μήκος μονοπατιού και η διασπορά του. Προτείνεται η χρήση της συνάρτησης  all_pairs_shortest_path_length του networkx.
#### B2.2 Να υπολογιστεί η διάμετρος (συνάρτηση diameter) και η ακτίνα (συνάρτηση radius).
#### Για τα παραπάνω πραγματοποιήστε συγκρίσεις και εξηγήστε με βάση τα χαρακτηριστικά της κάθε τοπολογίας όπως παρουσιάζονται στη θεωρία.

```python
#Για τη συνέχεια της άσκησης δημιουργούμε έναν νέο γράφο RGER χρησιμοποιώντας χαμηλότερη τιμή seedno ώστε ο γράφος να είναι connected. 
random.seed(500)
#n=4x 
n=49
M = 150

# για την κατασκευή του RGER χρησιμοποιήστε την gnm_random_graph από το networkx
RGER2=nx.gnm_random_graph(n, M)
RGER2.graph['name'] = 'RGER2'
```

REG: Έχει το μεγαλύτερο μέσο μήκος μικρότερου μονοπατιού και τις υψηλότερες τιμές διαμέτρου και ακτίνας όπως αναμέναμε, διότι ο κάθε κόμβος είναι συνδεδεμένος μόνο με τους διπλανούς του. 

RGER2: Σε σύγκριση με την τοπολογία REG, εδώ παρατηρούμε εμφανώς μικρότερες τιμές τόσο στο μέσο μήκος μονοπατιού όσο και στην διάμετρο και στην ακτίνα. Συνεπώς οι κόμβοι εδώ είναι πιο αποτελεσματικά συνδεδεμένοι.

SW: Παρά τις ομοιότητες με την SF λόγω κεντρικών κόμβων, παρουσιάζει μεγαλύτερο μήκος μονοπατιού, διάμετρο και ακτίνα όχι μόνο από την SF αλλά και από τις RGG,RGER2. Η πιθανότερη αιτία για αυτό είναι οι μεγάλες αποστάσεις στις συνδεσείς μεταξύ απομακρυσμένων κόμβων του γράφου από διάφορες περιοχές του.

#### B.2.3 Για κάθε τοπολογία, ποιος κόμβος συνδέεται με τους περισσότερους κόμβους με μήκος ελάχιστου μονοπατιού ίσο με τη διάμετρο της τοπολογίας?

```python
def paths_count(paths, diameter):
    count = 0
    for p in paths :
      if len(paths.get(p)) == diameter:
        count += 1
    return count

def find_nodes(G, diameter):
    return max(nx.all_pairs_shortest_path(G), key=lambda x : paths_count(x[1], diameter))

print('Node with most diameter shortest paths')
for g in Graphs2:
  if nx.is_connected(g):
    print(g.graph['name'] + ' : ')
    diameter = nx.diameter(g)
    node = find_nodes(g, diameter)
    print('Node label : ' + str(node[0]))
    print('Nodes : ' + str([x for x in node[1] if len(node[1].get(x)) == diameter]))
  else :
    print(g.graph['name'] + ' : Not Connected')
```

```python
# να προσδιοριστεί το label του κόμβου και το σύνολο των κόμβων 
# με τους οποίους συνδέεται με μήκος ελάχιστου μονοπατιού ΄ίσο με τη διάμετρο. 
```

#### Κέντρο του γράφου ορίζεται να είναι το σύνολο των κόμβων του που έχουν εκκεντρότητα ίση με την ακτίνα του.

```python
def find_periphery(G):
    return nx.periphery(G)
```

SW: Όπως και στην SF τοπολογία, οι κόμβοι με το μεγαλύτερο βαθμό αποτελούν το κέντρο του γράφου και εκείνοι με τον μικρότερο, που απαιτούν άρα περισσότερα βήματα για να φτάσουν σε άλλους κόμβους, ανήκουν στην περιφέρεια.

## Γ) Mετρικές κεντρικότητας κόμβων

### Γ1. Κεντρικότητα βαθμού (Degree centrality)
##### Γ1.1 Να αναπαρασταθεί η κατατανομή βαθμών κόμβων για κάθε τοπολογία.
##### Γ1.2 Να υπολογιστεί ο μέσος βαθμός κόμβου καθώς και η διασπορά των βαθμών κόμβου για κάθε τοπολογία.
##### Να σχολιαστούν τα αποτελέσματα.

```python
def find_degree(G):
    node_deg_centr = nx.degree_centrality(G)
    node_deg_centr_list = [node_deg_centr.get(x) for x in node_deg_centr]
    print('Mean value : ' + str(round(np.mean(node_deg_centr_list), 3)) +
          ', Variance : ' + str(round(np.var(node_deg_centr_list), 3)))
   
    plt.figure()
    plt.hist(node_deg_centr_list)
    plt.title('Node Degree Centrality')
    plt.xlabel('Degree Centrality')
    plt.ylabel('Nodes')
    plt.show()


for g in Graphs2:
    print(g.graph['name'] )
    find_degree(g)
```

REG: Παρατηρούμε ότι όλοι οι κόμβοι έχουν το ίδιο degree centrality. Αυτό συμβαίνει διότι έχουν όλοι τον ίδιο βαθμό αφού όλοι έχουν την ίδια απόσταση από τους άλλους κόμβους.

RGER: Η διασπορά είναι αρκετά χαμηλή και άρα οι περισσότεροι κόμβοι βρίσκονται σε κοντινές μεταξύ τους αποστάσεις.

RGG: Πολλοί κόμβοι έχουν μεγάλους βαθμούς, γι'αυτό και η μέση τιμή είναι υψηλή. Αυτό οφείλετααι στην μεγάλη ακτίνα που προκαλέι πολλές συνδέσεις.

SF: Παρατηρούμε ότι λίγοι κόμβοι έχουν μεγάλους βαθμούς ενώ οι περισσότεροι έχουν μικρούς και η διακύμανση είναι η μεγαλύτερη από όλες.

SW: Όλοι οι κόμβοι έχουν μικρή κεντρικότητα βαθμού κόμβου λόγω των μικρών παραμέτρων d,p που επιλέξαμε κατά την κατασκευή της τοπολογίας και η διασπορά είναι μηδενική. 

### Γ2. Κεντρικότητα εγγύτητας (Closeness centrality)
##### Γ2.1 Να αναπαρασταθεί η κατατανομή της εγγύτητας κόμβων για κάθε τοπολογία.

### Γ4. Kεντρικότητα Katz
#### Γ4.1 Να αναπαρασταθεί η κατανομή της κεντρικότητας Katz των κόμβων για κάθε τοπολογία. 
Για τον προσδιορισμό της παραμέτρου $alpha$ της συνάρτησης katz_centrality,  να υπολογιστεί η μεγαλύτερη ιδιοτιμή $λ_{max}$ του πίνακα γειτνίασης με τη συνάρτηση adjacency_spectrum του networkx. 
H $alpha$ ορίζεται να είναι ιση με $\frac{1}{λ_{max}}-0.01$.
#### Γ4.2 Να υπολογιστεί η μέση  κεντρικότητα Katz  καθώς και η διασπορά της για κάθε τοπολογία.
#### Να σχολιαστούν τα αποτελέσματα.

```python
def find_katz(G):
    l_max = np.real(max(nx.adjacency_spectrum(G)))
    alpha_value = 1 / l_max - 0.01
    node_deg_centr = nx.katz_centrality(G, alpha=alpha_value)
    node_deg_centr_list = [node_deg_centr.get(x) for x in node_deg_centr]
    print('Mean value : ' + str(round(np.mean(node_deg_centr_list), 3)) +
          ', Variance : ' + str(round(np.var(node_deg_centr_list), 3)))
   
    plt.figure()
    plt.hist(node_deg_centr_list,histtype='step')
    plt.title('Node Katz Centrality')
    plt.xlabel('Degree Centrality')
    plt.ylabel('Nodes')
    plt.show()


for g in Graphs2:
    print(g.graph['name'] )
    find_katz(g)
```

REG: Η διασπορά είναι μηδενική και η μέση τιμή της Katz κεντρικότητας ταυτίζεται με την τιμή του κάθε κόμβου αφού όλοι έχουν τα ίδια χαρακτηριστικά και τον ίδιο αριθμό γειτονικών  κόμβων.

RGER: Παρουσιάζει χαμηλότερη μέση τιμή και σχετικά μικρή διασπορά. Άρα οι κόμβοι εδώ έχουν μικρότερη συμμετοχή σε μονοπάτια μεταξύ άλλων κόμβων.

RGG: Παρατηρούμε ότι υπάρχουν κεντρικόι κόμβοι με μεγαλύτερη κεντρικότητα αλλά και πιο απομακρυσμένοι με μικρότερη.

SF: Εδώ η διασπορά είναι ανεβασμένη και έχουμε λίγους κεντρικούς κόμβους με μεγάλη επιρροή και άρα με μεγάλο βαθμό ενώ οι υπόλοιποι κόμβοι έχουν αρκετά μικρότερες τιμές.

#### Να προσδιοριστούν, για κάθε τοπολογία, οι πιο κεντρικοί κόμβοι ως προς το βαθμό, την εγγύτητα, την ενδιαμεσική κεντρικότητα και την κεντρικότητα Katz. Να σχολιαστούν τα αποτελέσματα.

```python
def find_central(G):
    node_centr = nx.degree_centrality(G)
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Degree Centrality : ' + str(node_max_list))

    node_centr = nx.closeness_centrality(G) 
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Closeness Centrality : ' + str(node_max_list))
    
    node_centr = nx.betweenness_centrality(G) 
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Betweeness Centrality : ' + str(node_max_list))

    l_max = np.real(max(nx.adjacency_spectrum(G))) 
    alpha_value = 1 / l_max - 0.01
    node_centr = nx.katz_centrality(G, alpha=alpha_value)
    max_n = node_centr.get(max(node_centr, key=lambda x: node_centr.get(x)))
    node_max_list = [x for x in node_centr if node_centr.get(x) == max_n]
    print('Katz Centrality : ' + str(node_max_list))

for g in Graphs2 : 
  print(g.graph['name'] + ' : ')
  find_central(g)
```

### Γ5.  Κεντρικότητα Pagerank 
#### Γ5.1 Να γίνει ανάλυση στο κατευθυνόμενο δίκτυο Web Google του οποίου οι ακμές δίνονται στο αρχείο web_Google.txt. Περισσότερες πληροφορίες για το dataset δίνονται στο link https://snap.stanford.edu/data/web-Google.html .
##### Να υπολογιστούν τα ακόλουθα
<ol>
<li>Ο αριθμός των κόμβων του γράφου</li>
<li>Ο αριθμός των κόμβων του γράφου με self-edge, δηλαδή ο αριθμός των κόμβων a ∈V  όπου (a,a) ∈Ε</li>
<li>Ο αριθμός των κατευθυνόμενων ακμών του γράφου, δηλαδή ο αριθμός των ταξινομημένων ζευγαριών  (a,b) ∈Ε με a≠b</li>
<li>Ο αριθμός των αμφίδρομων ακμών του γράφου, δηλαδή  ο αριθμός των μοναδικών μη ταξινομημένων ζευγαριών (a,b) ∈Ε με a≠b για τα οποία (a,b) ∈Ε και (b,a) ∈Ε.</li>
<li>Ο αριθμός των κόμβων με μηδενικό out-degree</li>
<li>Ο αριθμός των κόμβων με μηδενικό in-degree</li>
<li>Ο αριθμός των κόμβων με out-degree >10</li>
<li>Ο αριθμός των κόμβων με in-degree >10</li>
</ol>

```python
G_web = nx.read_edgelist("C:\\Users\\annak\\Downloads\\web-Google.txt", create_using=nx.DiGraph(), nodetype=int)

def analyze_graph(G):
    num_nodes = G.number_of_nodes()
    
    self_edges = [node for node in G.nodes if G.has_edge(node, node)]
    num_self_edges = len(self_edges)
    
    num_directed_edges = G.number_of_edges() - num_self_edges
    
    bidirectional_edges = sum(1 for u, v in G.edges if G.has_edge(v, u) and u != v) // 2
    
    zero_out_degree = sum(1 for node in G.nodes if G.out_degree(node) == 0)
    
    zero_in_degree = sum(1 for node in G.nodes if G.in_degree(node) == 0)
    
    out_degree_gt_10 = sum(1 for node in G.nodes if G.out_degree(node) > 10)
    
    in_degree_gt_10 = sum(1 for node in G.nodes if G.in_degree(node) > 10)
    
    print(f"Number of nodes: {num_nodes}")
    print(f"Number of nodes with self-edges: {num_self_edges}")
    print(f"Number of directed edges (excluding self-loops): {num_directed_edges}")
    print(f"Number of bidirectional edges: {bidirectional_edges}")
    print(f"Number of nodes with zero out-degree: {zero_out_degree}")
    print(f"Number of nodes with zero in-degree: {zero_in_degree}")
    print(f"Number of nodes with out-degree > 10: {out_degree_gt_10}")
    print(f"Number of nodes with in-degree > 10: {in_degree_gt_10}")

analyze_graph(G_web)
```

    </ol>

```python
weakly_connected_components = sorted(nx.weakly_connected_components(G_web), key=len, reverse=True)

if len(weakly_connected_components) < 6:
    raise ValueError("The graph has fewer than 6 connected components.")

component_nodes = weakly_connected_components[5]
subgraph = G_web.subgraph(component_nodes).copy()
pagerank_scores = nx.pagerank(subgraph)
plt.figure(figsize=(12, 12))
node_sizes = [pagerank_scores[node] * 10000 for node in subgraph.nodes]

nx.draw(subgraph, pos=nx.spring_layout(subgraph), node_size=node_sizes, 
        node_color="skyblue", edge_color="gray", with_labels=False, alpha=0.7)
plt.title("Subgraph (6th Largest Weakly Connected Component) with Node Sizes Proportional to PageRank Scores")
plt.show()

def plot_in_degree_distribution(G, title):
    in_degrees = [d for _, d in G.in_degree()]
    plt.figure(figsize=(8, 6))
    plt.hist(in_degrees, bins=np.logspace(0.1, 3, 50), color='skyblue', edgecolor='black', log=True)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel("In-degree")
    plt.ylabel("Frequency")
    plt.title(title)
    plt.grid(True, which="both", linestyle='--', linewidth=0.5)
    plt.show()

plot_in_degree_distribution(G_web, "In-Degree Distribution of the Full Graph")
plot_in_degree_distribution(subgraph, "In-Degree Distribution of the 6th Largest Component Subgraph")
```
